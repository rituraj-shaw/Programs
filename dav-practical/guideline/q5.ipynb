```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

load = load_iris()
df = pd.DataFrame(data=load.data, columns=load.feature_names)

# Creating the bar chart
plt.subplot(2, 2, 1)
plt.ylabel("Frequency")
plt.xlabel("Petal Length (cm)")
plt.bar(x=df['petal length (cm)'].value_counts().index, height=df['petal length (cm)'].value_counts())

plt.subplot(2, 2, 2)
plt.ylabel("Frequency")
plt.xlabel("Petal Width (cm)")
plt.bar(x=df['petal width (cm)'].value_counts().index, height=df['petal width (cm)'].value_counts())

plt.subplot(2, 2, 3)
plt.ylabel("Frequency")
plt.xlabel("Sepal Width (cm)")
plt.bar(x=df['sepal width (cm)'].value_counts().index, height=df['sepal width (cm)'].value_counts())

plt.subplot(2, 2, 4)
plt.ylabel("Frequency")
plt.xlabel("Sepal Length (cm)")
plt.bar(x=df['sepal length (cm)'].value_counts().index, height=df['sepal length (cm)'].value_counts())

plt.tight_layout()
plt.show()

# Show the plot
plt.show()
```
