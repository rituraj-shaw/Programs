```python
import pandas as pd
from datetime import datetime

dframe1 = pd.read_csv('weather_prediction_dataset.csv')
```

```python
dframe1
```

Part a: Compute mean of a series grouped by another series.

```python
mean_series = dframe1.groupby(['MONTH','BASEL_humidity'], as_index= False).mean()
mean_series
```

Part b: Fill an intermittent time series to replace all missing dates with values of previous non-missing date.

```python
dframe1['DATE'].fillna(method = 'ffill',inplace=True)
dframe1
```

Part c: Perform appropriate year-month string to dates conversion.

```python
dframe1['DATE'] = dframe1['DATE'].apply(lambda x: datetime.strptime(str(x), '%Y%m%d'))
dframe1
```

Part d: Split a dataset to group by two columns and then sort the aggregated results within the groups. 

```python
df_agg = dframe1.groupby(['MONTH','BASEL_cloud_cover']).agg({'BASEL_humidity':sum})
result = df_agg['BASEL_humidity'].groupby(level=0, group_keys=False)
print(result.nlargest())
```

Part e: Split a given dataframe into groups with bin counts.

```python
groups = dframe1.groupby(['MONTH', pd.cut(dframe1.BASEL_cloud_cover, 3)])
result = groups.size().unstack()
result
```
